```python
import scrapy
from scrapy.crawler import CrawlerProcess
from scrapy import Selector
```

```python
class CourseSpider(scrapy.Spider):
    
    name = 'Course_Spider'
    
    def start_requests(  self  ):
        for url in urls:
            yield scrapy.Request(url = url, callback = self.parse)
            
            
    def parse(  self, response  ):
        course_links = response.css('td.width > a::attr(href)').extract()
        for link in course_links:
            newlink = prefix+link
            yield response.follow(url = newlink, callback = self.parse_desc)
            
    
    def parse_desc(self, response):
        title = response.css('title::text').extract()
        description = response.css('h1#course_preview_title').xpath('../text()').extract()
        classtups.append((title, description))
        
```

```python
urlp1 = 'https://courses.cornell.edu/content.php?catoid=41&catoid=41&navoid=11626&filter[item_type]=3&filter[only_active]=1&filter[3]=1&filter[cpage]='
urlp2 = '#acalog_template_course_filter'

urls=[]
for i in range(1, 114):
    urls.append(urlp1 + str(i) + urlp2)

prefix = 'https://courses.cornell.edu/'
classtups = []

process = CrawlerProcess()
process.crawl(CourseSpider)
process.start()
```

```python
orderedclasstups = sorted(classtups, key = lambda x: str(x[0]).lower())
```

```python
import pandas as pd
import numpy as np
```

```python
df = pd.DataFrame(orderedclasstups, columns= ['Class Name', 'Class Description'])
```
